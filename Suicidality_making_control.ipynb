```python
import pandas as pd
phenotype = pd.read_csv('ABCD_phenotype_total - ABCD_phenotype_total.csv')
```

```python
total_CBCL = 'TotProb.CBCL'
ksads = phenotype.columns[73:114]
print(ksads)
```

```python
for k in ksads:
    print(k)
```

```python
phenotype.iloc[55][total_CBCL]
```

```python
control_subjects = []

for i in range(len(phenotype)):
    if phenotype.iloc[i][total_CBCL] <= 65:
        k_count = 0
        for k in ksads:
            if phenotype[k][i] != 0:
                k_count += 1
        if k_count == 0:
            control_subjects.append(phenotype.iloc[i]['subjectkey'])
```

```python
cs = pd.DataFrame(control_subjects)
cs.to_csv('control_subjects.csv')
```
