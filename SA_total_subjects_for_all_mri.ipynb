```python
import pandas as pd
import numpy as np
s_train = pd.read_csv('data_sa_unppc_sMRI_train.csv')
s_test = pd.read_csv('data_sa_unppc_sMRI_test.csv')
s = pd.concat([s_train, s_test])
print(s_train.shape, s_test.shape, s.shape)

d_train = pd.read_csv('data_sa_unppc_dMRI_train.csv')
d_test = pd.read_csv('data_sa_unppc_dMRI_test.csv')
d = pd.concat([d_train, d_test])

rsf_train = pd.read_csv('sa_ppc_rsfMRI_train.csv')
rsf_test = pd.read_csv('sa_ppc_rsfMRI_test.csv')
rsf = pd.concat([rsf_train, rsf_test])

n_train = pd.read_csv('sa_ppc_nbackfMRI_train.csv')
n_test = pd.read_csv('sa_ppc_nbackfMRI_test.csv')
n = pd.concat([n_train, n_test])

mid_train = pd.read_csv('sa_ppc_midfMRI_train.csv')
mid_test = pd.read_csv('sa_ppc_midfMRI_test.csv')
mid = pd.concat([mid_train, mid_test])

sst_train = pd.read_csv('sa_ppc_sstfMRI_train.csv')
sst_test = pd.read_csv('sa_ppc_sstfMRI_test.csv')
sst = pd.concat([sst_train, sst_test])

print(s.shape, d.shape, rsf.shape, n.shape, mid.shape, sst.shape)
```

```python
s_suid = s['SuicideAttempt.x'].values
d_suid = d['SuicideAttempt.x'].values
rsf_suid = rsf['SuicideAttempt'].values
n_suid = n['SuicideAttempt'].values
mid_suid = mid['SuicideAttempt'].values
sst_suid = sst['SuicideAttempt'].values
```

```python
suicide_sub = []
health_sub = []
for i in range(len(s_suid)):
    if s_suid[i] == 1:
        suicide_sub.append(s.iloc[i]['subjectkey'])
    else:
        health_sub.append(s.iloc[i]['subjectkey'])
        
for i in range(len(d_suid)):
    if d_suid[i] == 1:
        suicide_sub.append(d.iloc[i]['subjectkey'])
    else:
        health_sub.append(d.iloc[i]['subjectkey'])
        
for i in range(len(rsf_suid)):
    if rsf_suid[i] == 1:
        suicide_sub.append(rsf.iloc[i]['subjectkey'])
    else:
        health_sub.append(rsf.iloc[i]['subjectkey'])
        
for i in range(len(mid_suid)):
    if mid_suid[i] == 1:
        suicide_sub.append(mid.iloc[i]['subjectkey'])
    else:
        health_sub.append(mid.iloc[i]['subjectkey'])
        
for i in range(len(sst_suid)):
    if sst_suid[i] == 1:
        suicide_sub.append(sst.iloc[i]['subjectkey'])
    else:
        health_sub.append(sst.iloc[i]['subjectkey'])
```

```python
suicide_sub = set(suicide_sub)
health_sub = set(health_sub)
print(len(suicide_sub))
print(len(health_sub))
```

```python
(pd.DataFrame(list(suicide_sub))).to_csv('SA_suicide_sub.csv')
(pd.DataFrame(list(health_sub))).to_csv('SA_health_sub.csv')
```
